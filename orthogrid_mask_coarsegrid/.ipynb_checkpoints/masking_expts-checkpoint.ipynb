```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import pyfits
import numpy as np
```

```python
# read dophot catalog
cat = np.genfromtxt('02a57639o0328I.dph')
```

```python
hdulist = pyfits.open('02a57639o0328I.fits')
img = hdulist[0].data
hdulist.close()
```

```python
x0,y0 = 10500/2,10500/2
n = 800
inwind = (cat[:,5]-(y0-n/2) > 0) & (cat[:,5]-(y0-n/2) < n) & (cat[:,6]-(x0-n/2) > 0) & (cat[:,6]-(x0-n/2) < n)
```

```python
np.max(img[x0-n/2:x0+n/2,y0-n/2:y0+n/2])
```

```python
plt.figure(figsize=(18,10))
plt.imshow(img[x0-n/2:x0+n/2,y0-n/2:y0+n/2],vmin=200,vmax=500,interpolation='nearest')
plt.colorbar()

plt.scatter(cat[inwind,5]-(y0-n/2),cat[inwind,6]-(x0-n/2),300,facecolors='none',edgecolors='k',alpha=.4)

plt.xlim([0,n])
plt.ylim([0,n])
```

```python
plt.figure(figsize=(18,10))
plt.imshow(img[x0-n/2:x0+n/2,y0-n/2:y0+n/2]>10000,interpolation='nearest')
plt.colorbar()

plt.xlim([0,n])
plt.ylim([0,n])
```
