```python
fhdlabels = [l.split('/')[-1].split('.')[0] for l in commands.getoutput('ls '+analysis_root+'*.p').split('\n')]
print(len(fhdlabels))
```

```python
jds = [l.split('jd')[1].split('_')[0] for l in fhdlabels]
pointings = [l.split('pointing')[1].split('_')[0] for l in fhdlabels]

unique_jds = list(np.unique(jds))
unique_pointings = list(np.unique(pointings))
```

```python
mwa_images = []
for i in range(len(fhdlabels)):
    print('loading '+(str(i)))
    mwa_images.append(pickle.load( open( analysis_root+fhdlabels[i]+'.p', "rb" ) ))
```

```python
cols = 'rgbkymc'
stys = ['-','--',':','-.','o']
print(len(cols)*len(stys))

col_sequence = []
#lw_sequence = []

# generate sequence of color-styles and linewidths
for styi in range(len(stys)):
    for coli in range(len(cols)):
        col_sequence.append(cols[coli]+stys[styi])
        #lw_sequence.append(lws[lwi])

print(col_sequence)
#print(lw_sequence)
```

```python
plt.figure(figsize=(12,14))

nbins = 12
lmax = 2100

lbincenters, Cl_dirty0, bin_counts, bin_weights = mwa_utils.img2PS(mwa_images[i].dirty_xx_u0, mwa_images[i].weights_xx0, mwa_images[i].dtheta_rad,nbins,lmax,hann=False)

for i in range(len(fhdlabels)):
    col = col_sequence[unique_pointings.index(pointings[i])]
    lw = 1
    
    lbincenters, Cl_dirty0, bin_counts, bin_weights = mwa_utils.img2PS(mwa_images[i].dirty_xx_u0, mwa_images[i].weights_xx0, mwa_images[i].dtheta_rad,nbins,lmax,hann=False)
    lbincenters, Cl_res0, bin_counts, bin_weights =     mwa_utils.img2PS(mwa_images[i].dirty_xx_u0 - mwa_images[i].model_xx_u0, mwa_images[i].weights_xx0,mwa_images[i].dtheta_rad, nbins,lmax, hann=False)
    #lbincenters, Cl_dirty_diff, bin_counts, bin_weights =  mwa_utils.img2PS(mwa_images[i].dirty_xx_u0 - mwa_images[i].dirty_xx_u1, mwa_images[i].weights_xx0,mwa_images[i].dtheta_rad, nbins,lmax, hann=False)
    
    plt.loglog(lbincenters,Cl_dirty0*1.e6,col,linewidth=lw)
    plt.loglog(lbincenters,Cl_res0*1.e6,col,linewidth=lw)
    #plt.loglog(lbincenters,Cl_dirty_diff*1.e6,col,linewidth=lw)
    

# Adam's deep 32hr run
lbincenters, Cl_dirty0, bin_counts, bin_weights = mwa_utils.img2PS(mwa_deep_image.dirty_xx_u0, mwa_deep_image.weights_xx0, mwa_deep_image.dtheta_rad,nbins,lmax,hann=False)
lbincenters, Cl_res0, bin_counts, bin_weights =     mwa_utils.img2PS(mwa_deep_image.dirty_xx_u0 - mwa_deep_image.model_xx_u0, mwa_deep_image.weights_xx0,mwa_deep_image.dtheta_rad, nbins,lmax, hann=False)
plt.loglog(lbincenters,Cl_dirty0*1.e6,'k',linewidth=3)
plt.loglog(lbincenters,Cl_res0*1.e6,'k',linewidth=3,label='32hr')
    
for i in range(len(unique_pointings)):
    plt.loglog([1,2],[1,2],col_sequence[i],label=str(unique_pointings[i]))

matplotlib.rcParams.update({'font.size': 22})
plt.ylabel(r'$C_\ell$ (mK$^2$)')
plt.ylim([.5e1,1.e4])
plt.legend()
#plt.ylim([10,1.e3])
plt.xlim([1.e2,3.e3])
plt.legend(fontsize=14)
```
