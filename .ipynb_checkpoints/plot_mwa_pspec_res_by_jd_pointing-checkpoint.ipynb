```python
fhdlabels = [l.split('/')[-1].split('.')[0] for l in commands.getoutput('ls '+analysis_root+'*.p').split('\n')]
print(len(fhdlabels))
```

```python
jds = [l.split('jd')[1].split('_')[0] for l in fhdlabels]
pointings = [l.split('pointing')[1].split('_')[0] for l in fhdlabels]

unique_jds = list(np.unique(jds))
unique_pointings = list(np.unique(pointings))
```

```python
mwa_images = []
for i in range(len(fhdlabels)):
    print('loading '+(str(i)))
    mwa_images.append(pickle.load( open( analysis_root+fhdlabels[i]+'.p', "rb" ) ))
```

```python
mwa_multiday_images = []
fhdmultidaylabels = ['Combined_obs_1effectiveday_wedge_cut_plus_res_cut','Combined_obs_2day_wedge_cut_plus_res_cut','Combined_obs_3day_wedge_cut_plus_res_cut','Combined_obs_5day_wedge_cut_plus_res_cut','Combined_obs_11day_wedge_cut_plus_res_cut']

for i in range(len(fhdmultidaylabels[:2])):
    print('loading '+(str(i)))
    freq_averaged_cubedat0,freq_averaged_cubedat1 = mwa_utils.load_freq_averaged_odd_even_cubedat(fhdcubesroot,fhdmultidaylabels[i])
    mwa_multiday_image = mwa_utils.MWAImage(freq_averaged_cubedat0,freq_averaged_cubedat1, dtheta_amin=7, n=100)    
    pickle.dump( mwa_multiday_image, open( analysis_root+'../'+fhdmultidaylabels[i]+'.p', "wb" ) )

for i in range(len(fhdmultidaylabels)):
    print('loading '+(str(i)))
    mwa_multiday_images.append(pickle.load( open( analysis_root+'../'+fhdmultidaylabels[i]+'.p', "rb" ) ))
```

```python
cols = 'rgbkymc'
stys = ['-','--',':','-.','o']
print(len(cols)*len(stys))

col_sequence = []
#lw_sequence = []

# generate sequence of color-styles and linewidths
for styi in range(len(stys)):
    for coli in range(len(cols)):
        col_sequence.append(cols[coli]+stys[styi])
        #lw_sequence.append(lws[lwi])

print(col_sequence)
#print(lw_sequence)
```
