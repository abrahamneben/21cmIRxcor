```python
cubedir='recubesatdifferentredshifts/'
[::-1]
z_vals = np.float32(zstr_vals)
Nf_vals = [::-1]

cubes_ir = []
cubes_21 = []
rescale = 1.e20
for i in range(len(z_vals)):
    cubes_ir.append(rescale*np.reshape(np.fromfile(cubedir+'cone_Nz_'+str(Nf_vals[i])+'_Nx_48_Ilya_erg_z'+zstr_vals[i]+'_d4.8_dg0.1.dat',dtype=np.float32),(Nf_vals[i],48,48)))
    
    # per https://www.evernote.com/shard/s316/nl/2147483647/a7de741c-c90f-40a9-9fdc-072d49c77073/
    cubes_21.append(1/(1+z_vals[i])**2*rescale*np.reshape(np.fromfile(cubedir+'cone_Nz_'+str(Nf_vals[i])+'_Nx_48_Tb_erg_z'+zstr_vals[i]+'_d4.8_dg0.1.dat',dtype=np.float32),(Nf_vals[i],48,48)))
```

```python
# cube details the PDF saved in https://www.evernote.com/shard/s316/nl/2147483647/55920242-06f6-412d-84b9-c0ef6932a02e/
dtheta_deg = 0.1
df_21 = 100.e3
```

```python
kjy = 1.e-23
erg2kjypersr = 1.e-7/(1.e-2)**2/kjy/rescale

kB = 1.38e-23
c=3.e8
z = 8
lamda=c*(1+z)/1420.e6
erg2kelvin = 1.e-7/(1.e-2)**2*lamda**2/kB/rescale
```

```python
matplotlib.rcParams.update({'font.size': 16})

plt.figure(figsize=(25,12))

for i in range(len(z_vals)):
    plt.subplot(4,len(z_vals),i+1)
    plt.plot(100*np.mean(cubes_ir[i],axis=(1,2))*erg2kjypersr,'b')
    plt.title('IR mean (z='+str(zstr_vals[i])+')')
    #plt.xlabel('freq index')
    plt.ylabel('.01 * kJy/sr')
    plt.ylim([0,01.5])
    
    plt.subplot(4,len(z_vals),len(z_vals)+i+1)
    plt.plot(100*np.std(cubes_ir[i],axis=(1,2))*erg2kjypersr,'b')
    plt.title('IR std (z='+str(zstr_vals[i])+')')
    #plt.xlabel('freq index')
    plt.ylabel('.01 * kJy/sr')
    plt.ylim([0,01.5])
    

    plt.subplot(4,len(z_vals),2*len(z_vals)+i+1)
    plt.plot(np.mean(cubes_21[i],axis=(1,2))*erg2kelvin,'r')
    plt.title('21 mean (z='+str(zstr_vals[i])+')')
    plt.ylabel('Kelvin')
    #plt.xlabel('freq index')
    plt.ylim([0,.05])
    
    plt.subplot(4,len(z_vals),3*len(z_vals)+i+1)
    plt.plot(np.std(cubes_21[i],axis=(1,2))*erg2kelvin,'r')
    plt.title('21 std (z='+str(zstr_vals[i])+')')
    plt.ylabel('Kelvin')
    #plt.xlabel('freq index')
    plt.ylim([0,.02])

plt.tight_layout()
```
