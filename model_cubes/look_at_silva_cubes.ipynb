```python
cubedir='recubesatdifferentredshifts/'
zstr_vals = np.array(['7.0','7.5','8.0','8.5','9.0'])[::-1]
z_vals = np.float32(zstr_vals)
Nf_vals = np.array([112,99,88,79,72])[::-1]

cubes_ir = []
cubes_21 = []
rescale = 1.e20
for i in range(len(z_vals)):
    cubes_ir.append(rescale*np.reshape(np.fromfile(cubedir+'cone_Nz_'+str(Nf_vals[i])+'_Nx_48_Ilya_erg_z'+zstr_vals[i]+'_d4.8_dg0.1.dat',dtype=np.float32),(Nf_vals[i],48,48)))
    cubes_21.append(rescale*np.reshape(np.fromfile(cubedir+'cone_Nz_'+str(Nf_vals[i])+'_Nx_48_Tb_erg_z'+zstr_vals[i]+'_d4.8_dg0.1.dat',dtype=np.float32),(Nf_vals[i],48,48)))
```
